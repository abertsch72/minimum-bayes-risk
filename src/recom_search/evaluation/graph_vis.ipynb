```python
path = '/mnt/data1/jcxu/back_to_fact/df'
import os,pickle
import seaborn as sns
import matplotlib.pyplot as plt
def read_f(path,fname):
    with open(os.path.join(path, fname),'rb') as fd:
        data = pickle.load(fd)
    return data
```

```python
import numpy
fname = 'best_The Subsea Power Hub (SPH)250_25_3_5_0'
# fig, axes = plt.subplots(10, 2)

for i in range(10):
    data = read_f(path, f"{fname}_{i}.pkl")

    data = data.astype({'avg': numpy.float64, 'sum': numpy.float64,'len': numpy.int32})

    # sample = data.sample(n=5000)
    # f, ax = plt.subplots(figsize=(6.5, 6.5))
    # sns.scatterplot(x="avg", y="sum",data=data, ax=ax)
    # sns.scatterplot(x="len", y="sum",data=data, ax=ax)
    sns.displot(data=data,x="len", kind="kde",height=2)
    # sns.displot(data=data,x="sum", kind="kde")
    g = sns.jointplot(data=data,x="sum", y="avg", kind="kde")
    g.plot_joint(sns.scatterplot, s=100, alpha=.5)
    g.plot_marginals(sns.histplot, kde=True)
    print('-'*40)

data = read_f(path, f"{fname}.pkl")

data = data.astype({'avg': numpy.float64, 'sum': numpy.float64,'len': numpy.int32})

# sample = data.sample(n=5000)
# f, ax = plt.subplots(figsize=(6.5, 6.5))
# sns.scatterplot(x="avg", y="sum",data=data, ax=ax)
# sns.scatterplot(x="len", y="sum",data=data, ax=ax)
sns.displot(data=data,x="len", kind="kde",height=2)
# sns.displot(data=data,x="sum", kind="kde")
g = sns.jointplot(data=data,x="sum", y="avg", kind="kde")
g.plot_joint(sns.scatterplot, s=100, alpha=.5)
g.plot_marginals(sns.histplot, kde=True)
```
